```python
import pandas as pd
```

# Other Costs

```python
other_costs = pd.read_parquet('data/in/other_costs.parquet')
```

```python
other_costs['custo'] = other_costs['custo'].str.replace(',', '.').astype(float)
```

```python
other_costs['custo'] = round(other_costs['custo'] * 0.15, 2)
```

```python
other_costs.loc[0, 'descricao'] = 'Salário'
```

```python
other_costs
```

```python
other_costs.to_parquet('data/out/other_costs.parquet')
```

# Sales

```python
sales = pd.read_parquet('data/in/sales.parquet')
```

```python
ids = list(sales['id_venda'])
```

```python
for col in sales.columns:
    if (sales[col].dtype == 'object') and (',' in sales.loc[0, col]):
        sales[col] = sales[col].str.replace(',', '.')
        sales[col] = round(sales[col].astype(float) * 0.85, 2)
```

```python
sales.to_parquet('data/out/sales.parquet', index=False)
```

# Products per Sale

```python
pps = pd.read_parquet([
    'data/in/pps1.parquet',
    'data/in/pps2.parquet',
    'data/in/pps3.parquet',
    'data/in/pps4.parquet'
])
```

```python
pps = pps[pps['id_venda'].isin(ids)]
```

```python
pps['custo_producao'] = pps['custo_producao'].str.replace(',', '.').astype(float) * 0.65
```

```python
pps.to_parquet('data/out/products_per_sale.parquet')
```
